```python
from mpasviewer import scvtmesh
from earthcmap import escmap
import cartopy.crs as ccrs
```

```python
path2data = 'sample_data'

mpasd = scvtmesh(grid_file=f'{path2data}/4jorge/x1.655362.grid.nc', 
                 diag_list=f'{path2data}/4jorgeDA/an.2023-01-07_00.00.00.nc')
```

```python
mpasd.dataset()
```

```python
dta = mpasd.load()
```

```python
var1 = 'vegfra'
cmap, norm = escmap("gvf")
mpasd.show(dta, var_name = var1, crs=ccrs.Orthographic(central_longitude=-97, central_latitude=42), cmap = cmap, norm = norm, figsize = (10, 10))
```
