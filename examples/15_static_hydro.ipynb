```python
from mpasviewer import scvtmesh
from earthcmap import escmap
```

```python
path2data = 'sample_data'

mpasd = scvtmesh(grid_file=f'{path2data}/regional_testcase/CONUS.static.nc')
```

```python
mpasd.dataset()
```

```python
dta = mpasd.load()
```

```python
dta
```

```python
cmap, norm = escmap("terrain")
mpasd.show(dta, var_name = 'ter', cmap = cmap, norm = norm, figsize = (10, 10))
```

```python
cmap, norm = escmap("lalc")
mpasd.show(dta, var_name = 'ivgtyp', cmap = cmap, norm = norm, figsize = (10, 10))
```

```python
cmap, norm = escmap("soil_comet")
mpasd.show(dta, var_name='isltyp', cmap = cmap, norm = norm, figsize = (10, 10))
```

```python
cmap, norm = escmap("sfc_albedo")
mpasd.show(dta, var_name='snoalb', cmap = cmap, norm = norm, figsize = (10, 10))
```
